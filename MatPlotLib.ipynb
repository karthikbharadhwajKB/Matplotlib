```python
np.random.seed(555)
plt.figure(figsize=(15,7))
winter = np.random.randint(20,35,size=40)
summer =np.random.randint(36,49,size=40)
rainfall = np.random.randint(6,130,size=40)
plt.plot(winter)
plt.plot(summer)
plt.plot(rainfall)
#for title
plt.title("Temperature Line Chart")
# x-axis label
plt.xlabel("Number of Days")
#y-axis label
plt.ylabel("Temperture and rainfall in mm")
#Legend
plt.legend(['winter','summer','rainfall'],loc=0)
plt.show();
```

```python
np.random.seed(555)
plt.figure(figsize=(15,7))
low = np.random.randint(15,26,size=12)
high =np.random.randint(25,49,size=12)
plt.plot(low, linestyle="--",color="#0cb577",lw =4)
plt.plot(high, linestyle=":",color ="red",lw=3)
#for title
plt.title("Average High & Low temp in Hyderabad",color="white")
# x-axis label
plt.xlabel("Month")
#y-axis label
plt.ylabel("Temp in C")
#Legend
plt.legend(['low','high'],loc=0)
#To change x-axis and y-axis 
plt.xticks(np.arange(0,10,1))
plt.yticks(np.arange(0,50,5))
plt.show();
```

## 2. PieChart 

```python
plt.figure(figsize=(10,10))

values =[150,12,96,36,78,100]
labels =['India','Russia','America','Italy','Germany','Canada']
explode =(0.1,0,0,0,0,0)
plt.pie(values,labels=labels,counterclock=False,explode=explode,shadow=True,autopct = "%1.3f%%")
plt.title("Population Desinty per Km2")
plt.legend(labels,loc=0)
plt.show()
```

## 3. Scatter plot

```python
np.random.seed(123)

plt.figure(figsize=(10,10))
#generate a random float number using a numpy between any float range
wt = np.random.uniform(60.1,68.5,size=10)
#print(wt)
ht= np.random.uniform(124.3,168.6,size=10)

plt.scatter(wt,ht,c='blue',marker='o')
plt.xlabel('Weight',fontsize=16)
plt.ylabel('Height',fontsize=16)
plt.title('Scatter plot - Height vs Weight',fontsize=20)
plt.show()
```

```python
#height and #weight
np.random.seed(123)
plt.figure(figsize=(10,8))
#males' weight
weight_male = np.random.uniform(68.5,90.3,size = 40)
height_male = np.random.uniform(160,180,size=40)
#female's weight 
weight_female = np.random.uniform(35.5,65.8,size = 40)
height_female = np.random.uniform(150,170,size=40)
plt.scatter(height_male,weight_male,c="red",marker='s')
plt.scatter(height_female,weight_female,c='blue')

plt.title("Male vs Female (Height vs weight)")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.legend(['Male','Height'],loc=2,prop={'size':15})
plt.show()
```

# Box Plot 

```python
import matplotlib.pyplot as plt 


plt.figure(figsize=(10,10))
box_plot_data = [freshers,leadmanager,director,ceo]
box = plt.boxplot(box_plot_data,patch_artist=True,labels=['freshers','leadmanager','director','ceo'])

colors =['darkblue','red','yellow','green']

for p,col in zip(box['boxes'],colors):
    p.set_facecolor(col)
    
plt.show()
    
```

## Histogram 

```python
np.random.seed(123)
Values = np.random.randint(24,100,size=65)
plt.figure(figsize=(15,5))
plt.hist(Values,bins=7,histtype='bar',color='c',label='Test score Data',edgecolor="black")
plt.legend()
plt.title("Histogram of score")
plt.show()
```

```python
np.random.seed(123)
Values = np.random.randint(24,100,size=65)
plt.figure(figsize=(15,5))
plt.hist(Values,bins=7,histtype='step',align='mid',color='c',label='Test score Data')
plt.legend()
plt.title("Histogram of score")
plt.show()
```
